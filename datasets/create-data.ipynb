```python
import pickle 
```

```python
with open('./data.pkl', 'rb') as f:
    data = pickle.load(f)
    train_dataset, val_dataset, test_dataset, id2rel, rel2id = data
```

```python
print(type(train_dataset))
print(len(train_dataset))
print(type(train_dataset[0]))
print(len(train_dataset[0]))
```

```python
train_tokens_list = []
train_relation_list = []
train_text_list = []

for r in range(80):
    for i in range(len(train_dataset[r])):
        train_tokens_list.append(train_dataset[r][i]['tokens'])
        train_relation_list.append(train_dataset[r][i]['relation'])
        train_text_list.append(train_dataset[r][i]['text'])
```

```python
valid_tokens_list = []
valid_relation_list = []
valid_text_list = []

for r in range(80):
    for i in range(len(val_dataset[r])):
        valid_tokens_list.append(val_dataset[r][i]['tokens'])
        valid_relation_list.append(val_dataset[r][i]['relation'])
        valid_text_list.append(val_dataset[r][i]['text'])
```

```python
test_tokens_list = []
test_relation_list = []
test_text_list = []

for r in range(80):
    for i in range(len(test_dataset[r])):
        test_tokens_list.append(test_dataset[r][i]['tokens'])
        test_relation_list.append(test_dataset[r][i]['relation'])
        test_text_list.append(test_dataset[r][i]['text'])
```

```python
with open('train_test_flatten.pkl', 'wb') as fw:
    pickle.dump(
        ((train_tokens_list, train_relation_list, valid_text_list), 
        (valid_tokens_list, valid_relation_list, valid_text_list), 
        (test_tokens_list, test_relation_list, test_text_list),
        id2rel, rel2id),
        fw
    )
```
